```python
import os 
from glob import glob 
import shutil
img_count=0
data_path="/backup/RAW/DET/SynthText/SynthText/"
mat_path=os.path.join(data_path,"gt.mat")
img_path="/backup/RAW/DET/DATA/synth/images/"
gt_path ="/backup/RAW/DET/DATA/synth/gts/"
```

```python
import scipy.io
mat = scipy.io.loadmat(mat_path)
mat
```

```python
from tqdm.auto import tqdm
import cv2 
import matplotlib.pyplot as plt 
img_names=mat['imnames'][0]
bboxes=mat['wordBB'][0]
texts=mat['txt'][0]
FALSE_COUNT=0
for img_name,bbox,text in tqdm(zip(img_names,bboxes,texts),total=len(img_names)):
    # image --> copy
    src_path=os.path.join(data_path,img_name[0])
    # bbox
    xs,ys=bbox
    x1s,x2s,x3s,x4s=xs
    y1s,y2s,y3s,y4s=ys 
    # text
    _txts=[]
    for txt in text:
        txt=txt.replace("\n"," ")
        for _txt in txt.split(" "):
            if _txt.strip():
                _txts.append(_txt)
    # save
    try:
        _fmt=src_path.split(".")[-1]
        dst_path=os.path.join(img_path,f"{img_count}.{_fmt}")
        shutil.copy(src_path,dst_path)
        
        _gt_path=os.path.join(gt_path,f"{img_count}.txt")
        lines=[]
        for x1,y1,x2,y2,x3,y3,x4,y4,t in zip(x1s,y1s,x2s,y2s,x3s,y3s,x4s,y4s,_txts):
            lines.append(f"{int(x1)},{int(y1)},{int(x2)},{int(y2)},{int(x3)},{int(y3)},{int(x4)},{int(y4)},{t}\n")
        with open(_gt_path,'w') as f:
            for idx,line in enumerate(lines):
                if idx==len(lines)-1:
                    line=line.replace("\n","")
                f.write(line)
        img_count+=1
    except Exception as e:
        FALSE_COUNT+=1
print("False Count:",FALSE_COUNT)
```
